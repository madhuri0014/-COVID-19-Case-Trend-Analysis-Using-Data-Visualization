```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataset = pd.read_csv("nation_level_daily.csv")

# Aggregate total numbers for each category
total_confirmed = dataset['Daily Confirmed'].sum()
total_recovered = dataset['Daily Recovered'].sum()
total_deceased = dataset['Daily Deceased'].sum()

# Data for pie chart
values = [total_confirmed, total_recovered, total_deceased]
labels = ["Total Confirmed", "Total Recovered", "Total Deceased"]
colors = ['skyblue', 'lightgreen', 'salmon']

# Plot pie chart
plt.figure(figsize=(8,8))
plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("COVID-19 Cases Distribution", fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
dataset = pd.read_csv("nation_level_daily.csv")

# Summing totals for the entire dataset
total_confirmed = dataset['Daily Confirmed'].sum()
total_recovered = dataset['Daily Recovered'].sum()
total_deceased = dataset['Daily Deceased'].sum()

# Values and labels
values = [total_confirmed, total_recovered, total_deceased]
labels = ['Confirmed', 'Recovered', 'Deceased']
colors = ['skyblue', 'lightgreen', 'salmon']

# Create donut chart
plt.figure(figsize=(8,8))
plt.pie(values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width':0.4})

plt.title('Donut Chart of COVID-19 Cases', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
dataset = pd.read_csv("nation_level_daily.csv")

# Set Seaborn style
sns.set(style="whitegrid")

# Create scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(
    x=dataset['Daily Confirmed'],
    y=dataset['Daily Recovered'],
    hue=dataset['Daily Deceased'],  # Optional: color by another variable
    palette='coolwarm',
    size=dataset['Daily Deceased'], # Optional: size of points by deaths
    sizes=(20,200),
    alpha=0.7
)

plt.title('Scatter Plot of Daily Confirmed vs Daily Recovered Cases', fontsize=16)
plt.xlabel('Daily Confirmed', fontsize=12)
plt.ylabel('Daily Recovered', fontsize=12)
plt.legend(title='Daily Deceased')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
dataset = pd.read_csv("nation_level_daily.csv")

# Create a figure
plt.figure(figsize=(10,6))

# Use Matplotlib's fill_between to mimic violin
data = dataset['Daily Confirmed'].dropna()
density = pd.Series(data).plot.kde()  # KDE using pandas plot
plt.fill_between(density.get_lines()[0].get_xdata(), 0, density.get_lines()[0].get_ydata(), color='skyblue', alpha=0.6)

plt.title('Violin-like KDE Plot of Daily Confirmed Cases', fontsize=16)
plt.xlabel('Number of Cases', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots display inline in Colab
%matplotlib inline

# Load dataset
dataset = pd.read_csv("nation_level_daily.csv")

# Add default year for proper datetime conversion
dataset['Date'] = dataset['Date'] + ' 2026'
dataset['Date'] = pd.to_datetime(dataset['Date'], format="%d %B %Y", errors='coerce')

# Drop invalid dates
dataset = dataset.dropna(subset=['Date'])

# Optional: sort by date
dataset = dataset.sort_values('Date')

# Set Seaborn style
sns.set_style("whitegrid")

# Create figure
plt.figure(figsize=(14,7))

# Plot Daily Confirmed with smooth line
sns.lineplot(x='Date', y='Daily Confirmed', data=dataset, label='Daily Confirmed',
             color='skyblue', marker='o', linewidth=2)

# Plot Daily Recovered
sns.lineplot(x='Date', y='Daily Recovered', data=dataset, label='Daily Recovered',
             color='lightgreen', marker='x', linewidth=2)

# Plot Daily Deceased
sns.lineplot(x='Date', y='Daily Deceased', data=dataset, label='Daily Deceased',
             color='salmon', marker='s', linewidth=2)

# Titles and labels
plt.title('COVID-19 Daily Cases Over Time', fontsize=18, weight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Number of Cases', fontsize=14)

# Rotate x-axis labels for readability
plt.xticks(rotation=45)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.5)

# Legend
plt.legend(title='Case Type', fontsize=12, title_fontsize=13)

# Make layout tight
plt.tight_layout()

# Show plot
plt.show()
```
